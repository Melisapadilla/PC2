```python
#importamos y leemos el database

linkGit="https://github.com/Melisapadilla/PC2/raw/refs/heads/main/HDI.xlsx%20-%20Table%201.csv" 
data=pd.read_csv(linkGit)
data
```

```python
#Elimino las columnas con las que no voy a trabajar 
keep=['Back','Table 1. Human Development Index and its components ','','Unnamed: 4','Unnamed: 6','Unnamed: 8','Unnamed: 10','Unnamed: 12','Unnamed: 14']
dontKeep=set(data.columns.to_list())-set(keep)
Newdata=data.drop(columns=dontKeep)
Newdata
```

```python
#Cambio los nombres correctos a las columnas restantes
change={'Back':"HDI_rank",'Table1. Human Development Index and its components' : "Country", 'Unnamed: 2': "HDI", 'Unnamed: 4': "Life expectancy at birth", 'Unnamed: 6': "Expected years of schooling",'Unnamed: 8': "Mean years of schooling", 'Unnamed: 10':"Gross national income (GNI) per capita", 'Unnamed: 12': "GNI per capita rank minus HDI rank",'Unnamed: 14' : "HDI rank"}
Newdata.rename(columns=change,inplace=True)
Newdata
```

```python
#Luego elimino las filas iniciales que no son parte de la data
HDI= Newdata.drop(range(0,4))
HDI
```

```python
#Quedaría eliminar las ultimas filas de los últimos países e información 
HDINEW= HDI.drop(range(201,276))
HDINEW
```

```python
#Reemplazo todas las comas por puntos
HDICleaned= HDINEW.replace(',', '.', regex=True)
HDICleaned
```

```python
#eliminamos las comas en GNI
HDICleaned['Gross national income (GNI) per capita']=HDICleaned['Gross national income (GNI) per capita'].str.replace(".",'')
HDICleaned
```

```python
#elimino todas las NaN
HDICleaned = HDICleaned.fillna('')
HDICleaned
```

```python
#elimino la palabra HDI RANK que aún queda
HDICleaned.iloc[0, 0] = ''
HDICleaned
```

```python
#quitamos los espacios que pueda haber
objectCols=HDICleaned.select_dtypes(include=['object']).columns
objectCols
```

```python
HDICleaned = HDICleaned.loc[:,objectCols].apply(lambda col:col.str.strip())
HDICleaned
#con esto, podemos asumir limpio el database
```
